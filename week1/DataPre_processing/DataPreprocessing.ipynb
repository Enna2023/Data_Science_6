This code is using scikit-learn's GridSearchCV class to perform hyperparameter tuning on a decision tree classifier. It is searching for the best value of the max_depth hyperparameter to use in the classifier, by fitting the classifier to the training data and evaluating its performance using cross-validation.

Here is a breakdown of the code:

parameters = {'max_depth':range(3,20)}: This defines the hyperparameter grid that the grid search will search over. It will search for the best value of max_depth in the range from 3 to 20, inclusive.

clf = GridSearchCV(tree.DecisionTreeClassifier(), parameters, n_jobs=4,cv=5,scoring = 'roc_auc'): This creates an instance of the GridSearchCV class and specifies the following arguments:

tree.DecisionTreeClassifier(): the model to be trained, in this case a decision tree classifier.
parameters: the hyperparameter grid to search over.
n_jobs=4: the number of parallel jobs to run. This will train the model on 4 different cores at the same time.
cv=5: the number of folds to use for cross-validation. The model will be trained and evaluated 5 times, with each fold being used as the test set once and the rest of the data being used as the training set.
scoring='roc_auc': the scoring metric to use for evaluating the model. In this case, the metric is the area under the receiver operating characteristic curve (ROC AUC).
clf.fit(X=final_tr, y=yTrain): This fits the grid search object to the training data. The grid search will try out all possible combinations of hyperparameter values specified in the grid and evaluate each one using cross-validation.

dt = clf.best_estimator_: This assigns the best-performing decision tree from the grid search to a variable dt.

### Part 2: Handling Class Imbalance

**What is class imbalance?**
Class imbalance in classification problem is too less data points of one class compared to another class we are trying to predict. It leads to machine to learn too much of the dominant class and too less about the minority class!

Let us use data from step 2 (one hot encoded), use SMOTE and resample data\
**SMOTE: Synthetic Minority Oversampling Technique**\
We can update the example to first oversample the minority class to have 20 percent the number of examples of the majority class (here about 8k), then use random undersampling to reduce the number of examples in the majority class. But we are NOT undersampling majority class as this worked best for this case and we will see the results later!\
\
**How to apply SMOTE oversampling and undersampling technique?**

```python
#Libraries and functions to load for class imbalance
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
over = SMOTE(sampling_strategy=0.2,random_state=2) 
#under = RandomUnderSampler(sampling_strategy=0.2) : we are avoid under sampling of non fraud data, to avoid removing data: this worked well in this case!
steps = [('o', over)] # only do oversampling
pipeline = Pipeline(steps=steps)
X_res, y_res = pipeline.fit_resample(xTrain_dummy, yTrain)


print('Original dataset shape %s' % Counter(yTrain))
print('Resampled dataset shape %s' % Counter(y_res))
```

The SMOTE function is an oversampling method that generates synthetic samples for the minority class in order to balance the dataset. The sampling_strategy parameter specifies what proportion of the majority class should be used to generate synthetic samples for the minority class. The random_state parameter is used to set the random seed for reproducibility.

The RandomUnderSampler function is an undersampling method that randomly selects samples from the majority class to remove in order to balance the dataset. The sampling_strategy parameter specifies what proportion of the majority class should be removed.

The steps list specifies which resampling methods should be applied in the pipeline. In this case, only oversampling is being applied.

The pipeline object is created using the steps list, and it is then fit to the input data xTrain_dummy and labels yTrain using the fit_resample method. This returns the resampled dataset as X_res and y_res.

Finally, the original dataset shape and the resampled dataset shape are printed using the Counter function. This allows you to compare the class balance in the original dataset to the class balance in the resampled dataset.

**As you see in the above output, the dataset originally had just 1393 fraud cases and arodun 40k non fraud transactions.
Now thanks to oversampling via SMOTE, we have number of fraud and non-fraud cases in the ratio 1:5 and are relatively balanced information for model to learn
Now let us use SMOTE'd data for training a decsision tree classifier on it!**

```python
parameters = {'max_depth':range(3,20)}
clf = GridSearchCV(tree.DecisionTreeClassifier(), parameters, n_jobs=4,cv=5,scoring = 'roc_auc')
clf.fit(X=final_tr_res, y=y_res)
dt_smote = clf.best_estimator_ #final decision tree!
print (clf.best_score_, clf.best_params_)
```
